```python
# Remove unnecessary columns 'ID' and 'FIELD4'
df.drop(['ID', 'FIELD4'], axis=1, inplace=True)
```

```python
# Count the number of each job title
df['Job_Title'].value_counts()
```

```python
# Import re for text pattern matching and cleaning
import re
```

```python
# remove special characters
df['Job_Title']=df['Job_Title'].apply(lambda x: re.sub(r'[^a-zA-Z0-9\s]', ' ', x))
```

```python
df['Job_Title'].value_counts()
```

```python
# Check my job titles category based on my Job_title
df['job_title_sim'].value_counts()
```

```python
# Job_Title and job_title_sim columns for comparison
df[['Job_Title','job_title_sim']].head(101)
```

```python
# Create mapping for standardize job titles into meaningful categories
job_map = {
    'data scientist': 'Data Scientist',
    'other scientist': 'Other Scientist',
    'data engineer': 'Data Engineer',
    'machine learning engineer': 'ML Engineer',
    'analyst': 'Data Analyst',
    'data analitics': 'Data Analyst',
    'data modeler': 'Data Engineer',
    'Data scientist project manager': 'Manager',
    'director': 'Director',
    'na': 'Other'
}
```

```python
# job_categories from job_title_sim using pre_defined
df['Job_Category'] = df['job_title_sim'].map(job_map)
```

```python
# Count the number of job_category
df['Job_Category'].value_counts()
```

```python
# Again display columns and rows to understand data
df.head(100)
```

```python
# Convert seniority_title to lowercase and remove extra spaces
df['seniority_by_title'].str.lower().str.strip()
```

```python
# Standardize seniority_title into meaningful title
df['Seniority_Title'] = df['seniority_by_title'].replace({
    'jr': 'Junior',
    'sr': 'Senior',
    'na': 'Not Specified'
})
```

```python
# Count the number of Seniority_title
df['Seniority_Title'].value_counts()
```

```python
# Convert Degree column to string, remove extra spaces and lowercase with meaningful Degree name
df['Degree'] = df['Degree'].astype(str).str.strip().str.lower()

df['Degree_Required'] = df['Degree'].replace({
    'm': 'Masters',
    'p': 'PhD',
    'na': 'Not_Specified'
})
```

```python
# Count the number of Degree
df['Degree_Required'].value_counts()
```

```python
df.head(51)
```

```python
# Now Remove old columns
df.drop(['job_title_sim', 'seniority_by_title','Degree'], axis=1, inplace=True)
```

```python
# To see the Rating Column with count
df['Rating'].value_counts()
```

```python
# Import NumPy library for numerical calculation and handling missing values
import numpy as np
```

```python
# Replace invalid rating values (-1) with NaN to mark them as missing
df['Rating'].replace(-1, np.nan)
```

```python
# Create a new column 'Company_Rating' by filling missing Rating values with the median
df['Company_Rating'] = df['Rating'].fillna(df['Rating'].median())
```

```python
# To see all unique company names in the dataset
list(df['Company_Name'].unique())
```

```python
# Clean Company_Name by removing extra information (like ratings)
df['Company_Name'] = df['Company_Name'].str.split('\n').str[0]
```

```python
# Again check after removing extra info
list(df['Company_Name'].unique())
```

```python
# To see Location and Job_Location to compare both fields
df[['Location','Job_Location']].head(101)
```

```python
# Create new column 'State_Wise' and Extract state code from Location column
df['State_Wise'] = df['Location'].str.split(',').str[-1].str.strip()
```

```python
# To see all unique company_employee_size in the dataset
list(df['Size'].unique())
```

```python
# Create new column and Replace invalid size values ('Jan-50', 'unknown') with NaN
df['Company_Man_Power'] =df['Size'].replace(['Jan-50', 'unknown'], np.nan)
```

```python
# Fill missing values (NaN) with 'Unknown' for better analyzing
df['Company_Man_Power'] = df['Company_Man_Power'].fillna('Unknown')
```

```python
# After cleaning to see unique company_manpower_categories
list(df['Company_Man_Power'].unique())
```

```python
# Define a function to categorize company size into meaningful groups with new column
def size_category(x):
    if x == '51 - 200':
        return 'Small'
    elif x in ['201 - 500', '501 - 1000']:
        return 'Medium'
    elif x in ['1001 - 5000', '5001 - 10000']:
        return 'Large'
    elif x == '10000+':
        return 'Very Large'
    else:
        return 'Unknown'

df['Company_Man_Power_Category'] = df['Company_Man_Power'].apply(size_category)
```

```python
# Compare company man power and categorized size
df[['Company_Man_Power', 'Company_Man_Power_Category']].head(20)
```

```python
df.head(21)
```

```python
# After creating new column company_rating and company_size then drop the old column
df.drop(['Rating','Size'], axis=1, inplace=True)
```

```python
# Convert Founded column to numeric and Replace invalid founded year (-1) with NaN
pd.to_numeric(df['Founded'], errors='coerce')
df['Founded_Year'] = df['Founded'].replace(-1, np.nan)
```

```python
# To see unique founded_years after cleaning
list(df['Founded_Year'].unique())
```

```python
# Fill missing founded_years values with 'Unknown'
df['Founded_Year'].fillna('Unknown')
```

```python
# Calculate company_age (current year - founded year)
df['Company_Age'] = 2026 - df['Founded_Year']
```

```python
# To verify Founded_Year and Company_Age
df[['Founded_Year','Company_Age']].head(50)
```

```python
df.info()
```

```python
# Convert Founded_Year to nullable integer type for clean format and to keep missing values
df['Founded_Year'] = df['Founded_Year'].astype('Int64')
```

```python
# Convert Company_Age to nullable integer type for better analysis
df['Company_Age'] = df['Company_Age'].astype('Int64')
```

```python
# After creating new column Founded_Year and company_age then drop the old column
df.drop(['Founded','Age'], axis=1, inplace=True)
```

```python
# To see unique ownership_types to understand data variations
list(df['Type_of_ownership'].unique())
```

```python
# To standardize ownership types into clear categories and Apply mapping to create a new cleaned ownership_type column
ownership_map = {
    'Company - Private': 'Private',
    'Company - Public': 'Public',
    'Government': 'Government',
    'Nonprofit Organization': 'Nonprofit',
    'Hospital': 'Healthcare',
    'College / University': 'Education',
    'School / School District': 'Education',
    'Subsidiary or Business Segment': 'Subsidiary',
    'Other Organization': 'Other'
}

df['Ownership_Type'] = df['Type_of_ownership'].map(ownership_map)
```

```python
# Check unique values after cleaning
list(df['Ownership_Type'].unique())
```

```python
# Check unique values
list(df['Industry'].unique())
```

```python
# Clean Industry column convert to string, remove spaces, and replace invalid '-1' with 'Unknown'
df['Industry'] = df['Industry'].astype(str).str.strip().replace('-1', 'Unknown')
```

```python
# Check unique values after cleaning
list(df['Industry'].unique())
```

```python
# Check unique values
list(df['Sector'].unique())
```

```python
# List of all skill-related columns for processing and Convert all skill columns to numeric
skill_cols = ['Python','spark','aws','excel','sql_','sas','keras','pytorch',
              'scikit','tensor','hadoop','tableau','bi',
              'flink','mongo','google_an']

df[skill_cols] = df[skill_cols].apply(pd.to_numeric, errors='coerce')
```
