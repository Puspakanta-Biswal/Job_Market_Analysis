## 1. States with Most Number of Jobs

```python
state_jobs = df['State_Wise'].value_counts()
```

```python
state_jobs.head(10)
```

## 2. Average Minimal and Maximal Salaries in Different States

```python
state_salary = df.groupby('State_Wise')[['Min_Salary','Max_Salary']].mean().round(2)
state_salary.sort_values(by='Max_Salary', ascending=False).head(10)
```

## 3. Average Salary in Different States

```python
Avg_Salary_State = df.groupby('State_Wise')['Avg_Salary'].mean().round(2)
Avg_Salary_State.sort_values(ascending=False).head(15)
```

## 4. Top 5 Industries with Maximum Job Postings

```python
Top_Industries = df['Industry'].value_counts().head(5)
Top_Industries
```

## 5. Companies with Maximum Job Openings

```python
Top_Companies = df['Company_Name'].value_counts().head(15)
Top_Companies
```

## 6. Job Titles with Most Number of Jobs

```python
Top_Jobs = df['Job_Category'].value_counts()
Top_Jobs
```

## 7. Salary of Job Titles with Most Number of Jobs

```python
Job_Salary = df.groupby('Job_Category')['Avg_Salary'].mean().round(2)
Job_Salary.sort_values(ascending=False)
```

## 8. Skills Required by Companies for Each Job Title

```python
skill_cols = ['Python','Spark','AWS','Excel','SQL','SAS','Keras','PyTorch',
              'Scikit_Learn','TensorFlow','Hadoop','Tableau','Power_BI',
              'Flink','MongoDB','Google_Analytics']
```

```python
skills_percent = df.groupby('Job_Category')[skill_cols].mean() * 100
skills_percent
```

## 9. Relation between Average Salary and Education

```python
Degree_Salary = df.groupby('Degree_Required')['Avg_Salary'].mean().round(2)
Degree_Salary.sort_values(ascending=False)
```

## 10. Important insights

1. Python,Excel,SQL are most in-demand skills.
2. CA State and MA State have highest job opportunities.
3. Data Scientist roles offer highest average salary.
4. Larger companies offer higher salaries.
5. Higher education (PhD) correlates with higher salary.
